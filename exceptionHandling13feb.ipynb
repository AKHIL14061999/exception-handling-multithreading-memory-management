```python
# Q2. Write a python program to print Python Exception Hierarchy.

# Ans: In Python, all exceptions are derived from the base class Exception. Here is an example Python program to print the exception hierarchy:
    
def print_exception_hierarchy(exception, level=0):
    print("|  " * level + "+--" + exception.__name__)
    for subclass in exception.__subclasses__():
        print_exception_hierarchy(subclass, level + 1)

# Print the exception hierarchy
print("Python Exception Hierarchy:")
print_exception_hierarchy(BaseException)
```

```python
# Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.


# Ans: In python "ArithmaticError" is a built-in class that is raised when a arithmatic operation fails. It is the base class
#       of other exception classes which is related to arithmatic errors.
#     Here are some of the errors that are defined in the ArithmeticError class:
    
#     1 ZeroDivisionError : ZeroDivisionError is raised when a number is divided by zero.
    
#     Example: In this example, we have defined two variables a and b, and we are trying to divide a by b. Since b is zero, 
#              the ZeroDivisionError exception is raised, and the program prints an error message "Error: division by zero".



a = 10
b = 0

try:
    c = a / b
except ZeroDivisionError:
    print("Error: division by zero")

    
    
#     2 OverflowError : This exception is raised when an arithmetic operation results in a number that is too large to be represented.
    
#     Example:  In this example, we are trying to multiply a number n that is equal to the maximum value that can be represented by a 
#               variable of type int in Python, plus one. Since n * n is too large to be represented as an integer, the OverflowError 
#               exception is raised, and the program prints an error message "Error: overflow".

            
import sys

try:
    n = sys.maxsize + 1
    m = n * n
except OverflowError:
    print("Error: overflow")
    
```

```python
# Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

# Ans : LookupError: LookupError is a subclass of "Exception" class and base class of saveral built-classes including 
#                 "keyError" and "IndexError". It is used as the base class for exception that occur when a key or index
#                  is not found in collecion or mapping object. by using lookuperror class we can catch both types of errors
#                    using a single except block.
        
#     1. KeyError: It is the subclass of LookupError class and it raise an exception when trying to non-existing key in a dictionary
        
#         for example: In this example In the example above, we first try to access the key "pear" in the dictionary my_dict.
#                     Since "pear" is not a key in the dictionary, a KeyError is raised. We catch the exception using a try-except
#                     block and print the error message.
            
my_dict = {"apple": 1, "banana": 2, "orange": 3}
try:
    value = my_dict["pear"]
except KeyError as e:
    print(f"Error key not found : {e}")

#          2. Indexerror: It is the subclass of LookupError class and it raise an exception when trying to acces non-existing index in a
#                        list or other sequnce objct
         
#         for example:  we try to access the element at index 3 in the list my_list. Since my_list only has three elements 
#                       (with indices 0, 1, and 2), trying to access index 3 will raise an IndexError. We catch the exception  
#                       using a try-except block and print the error message.
            
my_list = [1, 2, 3]
try:
    value = my_list[3]
except IndexError as e:
    print(f"Error: {e}")
```
